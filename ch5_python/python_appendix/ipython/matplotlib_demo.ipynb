```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# 1D horozontal plot
plt.hlines(1,1,20)  # Draw a horizontal line
knots = [1,3,9,15,17,18,20]
plt.xlim(0,21)
plt.ylim(0.5,1.5)

y = np.ones(np.shape(knots))   # Make all y values the same
ax = plt.gca()
ax.plot(knots,y,marker ='o', color = 'black')  
ax.axes.yaxis.set_visible(False)
plt.savefig('1D_plot.png')
plt.show()
```

```python
# A simple linear Line
y = np.linspace(0, 10, 20)  # 20 points with y-axis range from 0 to 10
plt.plot(y)
plt.title('This is a Title' , fontsize = 14)     
plt.xlabel('This is index', fontsize = 12)            
plt.ylabel('This is Y-values' , fontsize = 12)
```

```python
y= np.linspace(0, 10, 20)  # 20 points with y-axis range from 0 to 10
y2 = 1/2 * y +1 
plt.plot(y, linestyle= '-.',label = 'Line 1')
plt.plot(y+1,':', linewidth = 2, alpha=0.8, label = 'Line 2')
plt.plot(y+2,'-', linewidth = 3, alpha=0.6,label = 'Line 3')
plt.plot(y+3,'--', linewidth = 4,label = 'Line 4')
# Lienstyle: 
# '-'       solid line style
# '--'      dashed line style
# '-.'      dash-dot line style
# ':'       dotted line style
#Lienwidth controls for the width of aline.
plt.title('Different styles of lines' , fontsize = 14)     
plt. xlabel('This is index', fontsize = 12)            
plt.ylabel('This is Y-values' , fontsize = 12)
plt.legend(loc = 'best')
# loc is the location of legend: 
# 'best'         : 0, (only implemented for axes legends)
# 'upper right'  : 1,
# 'upper left'   : 2,
# 'lower left'   : 3,
# 'lower right'  : 4,
# 'right'        : 5,
# 'center left'  : 6,
# 'center right' : 7,
# 'lower center' : 8,
# 'upper center' : 9,
# 'center'       : 10,
```

```python
y3 = [i*2-5 for i in y]
plt.plot(y, linestyle= '-.',label = 'Line 1')
plt.plot(y3, label = 'Line 2')
plt.text (3,7, 'Two lines intersect here!', fontsize =10, color ='green')
plt.arrow(4,6.5, 4,-1,head_width = 0.2,ec ='green' )
plt.text (10,-3, '(10,-3)', fontsize =10, color ='black')
plt.title('Two lines',fontsize = 14)
plt.arrow(10,-4, 8,1,head_width = 0.2,ec ='black' )
plt. xlabel('This is index', fontsize = 12)
plt.ylabel('This is Y-values' , fontsize = 12)
plt.legend(loc = 'best')
```

```python
y = np.linspace(0,5,20)
y2 = [i**2-10 for i in y]
```

```python
fig, (plt1, plt2) = plt.subplots(2,1)
plt.subplots_adjust(hspace=0.5)
plt1.plot(y)
plt2.plot(y2)
plt1.set_title('First plot')
plt2.set_title('Second plot')
```

```python
fig, (plt1, plt2) = plt.subplots(1,2)
plt.subplots_adjust(wspace=0.5)
plt1.plot(y)
plt2.plot(y2)
plt1.set_title('First plot')
plt2.set_title('Second plot')
```

```python
fig2, ((plt1, plt2),(plt3, plt4)) = plt.subplots(2,2) 
plt1.plot(y2,y, linestyle='-.',linewidth=3,color = 'sienna')
plt2.plot(y,y2, linewidth=5,color = 'peachpuff')
plt4.scatter(y,y2, marker='^',color = 'gold')
plt3.scatter(y2,y, marker = 'p',color = 'palevioletred')
```

```python
import matplotlib.colors as mcolors
num_color = 2
xx, yy = np.meshgrid(np.arange(0,10, step=0.5),
                     np.arange(0,10, step=0.5))

Z = np.random.randint(num_color , size=xx.size) # Randomly generate integer: 0 or 1
Z= Z.reshape(xx.shape)
colors =  ['sienna','peachpuff']
plt.pcolormesh(xx, yy, Z, cmap=mcolors.ListedColormap(colors),shading='auto')
```

```python
plt.scatter(xx, yy, c=Z, cmap  = mcolors.ListedColormap(colors))
```
